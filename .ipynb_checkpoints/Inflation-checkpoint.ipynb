### Imputación con Mediana a valores NaN

```python
print(inflation.isna().sum())
```

```python
import pandas as pd

# Crear una copia para trabajar con la imputación
inflation_imputed = inflation.copy()

# Identificar las columnas problemáticas
non_positive_columns = inflation_imputed.loc[:, inflation_imputed.columns[2:]].apply(lambda x: (x <= 0).any())
problematic_columns = non_positive_columns[non_positive_columns].index.tolist()

# Establecer un valor positivo mínimo para imputar
min_positive_value = 0.01

# Imputar valores no positivos
for column in problematic_columns:
    inflation_imputed[column] = inflation_imputed[column].apply(lambda x: min_positive_value if x <= 0 else x)

# Comprobación que no hay valores no positivos en las columnas problemáticas
assert all(inflation_imputed[problematic_columns].min() > 0)
```

```python
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.stattools import adfuller

# Imputación con Medianas
for column in inflation_imputed.columns[2:]:
    median_value = inflation_imputed[column].median()
    inflation_imputed[column] = inflation_imputed[column].fillna(median_value)

# Estandarización
scaler = StandardScaler()
inflation_imputed.iloc[:, 2:] = scaler.fit_transform(inflation_imputed.iloc[:, 2:])

# Estacionaridad año 2022
adf_result = adfuller(inflation_imputed['2022'].dropna())
adf_result

inflation_imputed.head(), adf_result
```

```python
years_columns = inflation_imputed.columns[2:]
adf_results = {}

# ADF por año
for column in years_columns:
    adf_result = adfuller(inflation_imputed[column].dropna())
    adf_results[column] = {
        'ADF Statistic': adf_result[0],
        'p-value': adf_result[1],
        'Critical Values': adf_result[4],
        '1%': adf_result[4]['1%'],  # Critical value for a 99% confidence level
        '5%': adf_result[4]['5%'],  # Critical value for a 95% confidence level
        '10%': adf_result[4]['10%']  # Critical value for a 90% confidence level
    }

{year: adf_results[year] for year in list(years_columns)[:63]}  # Ver los primeros 5 años.
```

### World | Regions | Countries

```python
import pandas as pd

# Cargar el DataFrame principal
main_inflation = pd.read_csv('main_inflation.csv')

regions_list = [
    "Africa Eastern and Southern", "Africa Western and Central",
    "Central Europe and the Baltics", "East Asia & Pacific",
    "Europe & Central Asia", "European Union", "Latin America & Caribbean",
    "Middle East & North Africa", "North America", "OECD members",
    "Sub-Saharan Africa (excluding high income)", "South Asia (IDA & IBRD)"
]

# 1) DataFrame por países
inflation_countries = main_inflation[~main_inflation['Country Name'].isin(regions_list + ['World'])]

# 2) DataFrame por regiones
inflation_regions = main_inflation[main_inflation['Country Name'].isin(regions_list)]

# 3) DataFrame global
inflation_world = main_inflation[main_inflation['Country Name'] == 'World']

print(inflation_countries.head())
print(inflation_regions.head())
print(inflation_world.head())
```

### Visualización de Tendencias Temporales | Countries

```python
import matplotlib.pyplot as plt
import seaborn as sns

countries_to_plot = ['United States', 'Germany', 'Brazil']
data_to_plot = inflation_countries[inflation_countries['Country Name'].isin(countries_to_plot)]

plt.figure(figsize=(20, 12))
for country in countries_to_plot:
    subset = data_to_plot[data_to_plot['Country Name'] == country]
    plt.plot(subset.columns[2:], subset.iloc[0, 2:], label=country)  

plt.title('Final Inflation, consumer prices (annual %) Over Time')
plt.xlabel('Year')
plt.ylabel('Standardized Inflation (%)')
plt.legend()
plt.grid(True)
plt.show()
```

### Comparación entre Regiones

```python
regions_data_2021 = inflation_regions[['Country Name', '2021']] 

plt.figure(figsize=(12, 8))
sns.barplot(x='2021', y='Country Name', data=regions_data_2021)
plt.title('Inflation, consumer prices (annual %) by Region in 2021')
plt.xlabel('Standardized Inflation (%)')
plt.ylabel('Region')
plt.show()
```

### Tendencia Mundial

```python
plt.figure(figsize=(22, 6))
plt.plot(inflation_world.columns[2:], inflation_world.iloc[0, 2:], marker='o')  
plt.title('World Inflation, consumer prices (annual %) Over Time')
plt.xlabel('Year')
plt.ylabel('Standardized Inflation (%)')
plt.grid(True)
plt.show()
```

# Será lento

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

years = [str(year) for year in range(2014, 2023)]
inflation_filtered = inflation_regions.set_index('Country Name')[years].transpose()

# Resumen estadístico
print(inflation_filtered.describe())

# Gráficos de series temporales para cada país
for column in inflation_filtered.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(inflation_filtered.index, inflation_filtered[column])
    plt.title(f'Inflation, consumer prices (annual %) for {column}')
    plt.xlabel('Año')
    plt.ylabel('Final Inflation (%)')
    plt.grid(True)
    plt.show()

# Descomposición de series temporales del total agregado
result = seasonal_decompose(inflation_filtered.sum(axis=1), model='additive', period=1)
result.plot()
plt.show()

# Correlación entre regiones
plt.figure(figsize=(10, 8))
sns.heatmap(inflation_filtered.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación entre Regiones')
plt.show()
```
