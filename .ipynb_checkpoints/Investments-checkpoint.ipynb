# Gross Fixed Capital Formation (constant 2015 US$)
Formación Bruta de Capital Fijo (Gross Fixed Capital Formation, GFCF): Incluye todas las adquisiciones netas de activos fijos (tanto reemplazos como adiciones) más ciertos adiciones a los activos no producidos.

### Imputación con Mediana a valores NaN

```python
import pandas as pd
import numpy as np

# Crear una copia para trabajar con la imputación
investments_imputed = investments.copy()

# Convertir columnas de años a numérico y manejar errores en la copia
years_columns = investments_imputed.columns[2:]
for column in years_columns:
    investments_imputed[column] = pd.to_numeric(investments_imputed[column], errors='coerce')

# Calcular la mediana para cada columna de año en la copia
median_values = investments_imputed[years_columns].median()

# Imputar los valores NaN con la mediana correspondiente en la copia
for column in years_columns:
    median_val = median_values[column]
    if median_val <= 0:  # Ajuste si la mediana es cero o negativa
        median_val = investments_imputed[investments_imputed[column] > 0][column].median()  # Recalcular la mediana excluyendo ceros y negativos
    investments_imputed[column].fillna(median_val, inplace=True)

assert (investments_imputed[years_columns] <= 0).sum().sum() == 0, "Existen valores no positivos después de la imputación."

print(investments_imputed[years_columns].describe())
```

```python
# Verificar cuáles columnas todavía tienen valores no positivos y cuántos son
non_positive_counts = (investments_imputed[years_columns] <= 0).sum()
non_positive_columns = non_positive_counts[non_positive_counts > 0]

if not non_positive_columns.empty:
    print("Columnas con valores no positivos y su cantidad:")
    print(non_positive_columns)
else:
    print("No hay valores no positivos, la imputación ha sido exitosa.")
```

```python
# Asegurarse de que la mediana calculada es siempre positiva
for column in years_columns:
    positive_values = investments_imputed[column][investments_imputed[column] > 0]
    median_val = positive_values.median() if not positive_values.empty else 0.01

    # Imputar los valores NaN y los valores no positivos
    investments_imputed[column] = investments_imputed[column].apply(lambda x: median_val if x <= 0 or pd.isna(x) else x)

# Verificación final para asegurarse de que no hay valores no positivos
non_positive_counts = (investments_imputed[years_columns] <= 0).sum()
non_positive_columns = non_positive_counts[non_positive_counts > 0]
if not non_positive_columns.empty:
    print("Columnas con valores no positivos y su cantidad:")
    print(non_positive_columns)
else:
    print("No hay valores no positivos, la imputación ha sido exitosa.")
```

### Estandarización de las Variables

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

years_columns = investments_imputed.columns[2:]  

# Crear el objeto StandardScaler
scaler = StandardScaler()

# Ajustar y transformar los datos
investments_imputed[years_columns] = scaler.fit_transform(investments_imputed[years_columns])

means = investments_imputed[years_columns].mean(axis=0)  # Media debería ser cercana a 0
std_devs = investments_imputed[years_columns].std(axis=0)  # Desviación estándar debería ser 1

print("Medias de las columnas estandarizadas después de la imputación:\n", means)
print("\nDesviaciones estándar de las columnas estandarizadas después de la imputación:\n", std_devs)
```

### Verificación de Estacionariedad
Prueba Dickey-Fuller Aumentada (ADF):

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

years_columns = investments_imputed.columns[2:]  

adf_results = {}

for column in years_columns:
    adf_result = adfuller(investments_imputed[column].dropna())
    adf_results[column] = {
        'ADF Statistic': adf_result[0],
        'p-value': adf_result[1],
        'Critical Values': adf_result[4],
        '1%': adf_result[4]['1%'],  
        '5%': adf_result[4]['5%'],  
        '10%': adf_result[4]['10%']  
    }

adf_results_df = pd.DataFrame(adf_results).transpose()

print(adf_results_df)
```

### World | Regions | Countries

```python
import pandas as pd

# Cargar el DataFrame principal
main_investments = pd.read_csv('main_investments.csv')

# Lista de regiones y World
regions_list = [
    "Africa Eastern and Southern", "Africa Western and Central",
    "Central Europe and the Baltics", "East Asia & Pacific",
    "Europe & Central Asia", "European Union", "Latin America & Caribbean",
    "Middle East & North Africa", "North America", "OECD members",
    "Sub-Saharan Africa (excluding high income)", "South Asia (IDA & IBRD)"
]

# 1) DataFrame para países
investments_countries = main_investments[~main_investments['Country Name'].isin(regions_list + ['World'])]

# 2) DataFrame para regiones
investments_regions = main_investments[main_investments['Country Name'].isin(regions_list)]

# 3) DataFrame para World
investments_world = main_investments[main_investments['Country Name'] == 'World']

print(investments_countries.head())
print(investments_regions.head())
print(investments_world.head())
```

### Visualización de Tendencias Temporales | Countries

```python
import matplotlib.pyplot as plt
import seaborn as sns

countries_to_plot = ['United States', 'Germany', 'Korea, Rep.', 'United Kingdom']
data_to_plot = investments_countries[investments_countries['Country Name'].isin(countries_to_plot)]

plt.figure(figsize=(20, 12))
for country in countries_to_plot:
    subset = data_to_plot[data_to_plot['Country Name'] == country]
    plt.plot(subset.columns[2:], subset.iloc[0, 2:], label=country)  

plt.title('Final Gross Fixed Capital Formation (constant 2015 US$) Over Time')
plt.xlabel('Year')
plt.ylabel('Standardized Gross Fixed Capital Formation')
plt.legend()
plt.grid(True)
plt.show()
```

### Comparación entre Regiones

```python
regions_data_2009 = investments_regions[['Country Name', '2009']]  

plt.figure(figsize=(12, 8))
sns.barplot(x='2009', y='Country Name', data=regions_data_2009)
plt.title('Gross Fixed Capital Formation (constant 2015 US$) by Region in 2009')
plt.xlabel('Standardized Gross Fixed Capital Formation')
plt.ylabel('Region')
plt.show()
```

### Tendencia Mundial

# Será lento

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

years = [str(year) for year in range(2014, 2023)]
data_filtered = investments_regions.set_index('Country Name')[years].transpose()

# Resumen estadístico
print(investments_filtered.describe())

# Gráficos de series temporales para cada país
for column in data_filtered.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(data_filtered.index, data_filtered[column])
    plt.title(f'Gross Fixed Capital Formation (constant 2015 US$) for {column}')
    plt.xlabel('Año')
    plt.ylabel('Final Gross Fixed Capital Formation')
    plt.grid(True)
    plt.show()

# Descomposición de series temporales del total agregado
result = seasonal_decompose(data_filtered.sum(axis=1), model='additive', period=1)
result.plot()
plt.show()

# Correlación entre regiones
plt.figure(figsize=(10, 8))
sns.heatmap(data_filtered.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación entre Regiones')
plt.show()
```
