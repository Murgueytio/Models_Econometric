## Evaluando, reflexionando y eligiendo desde el OpenSource del World Bank el indicador adecuado para mi modelo econométrico.

- **Balanza comercial**: Este es el indicador más directo del balance comercial y mide la diferencia entre el valor de las exportaciones y las importaciones de bienes y servicios de un país, se relaciona directamente con el rendimiento económico y se usa frecuentemente en estudios macroeconómicos para evaluar su impacto sobre el crecimiento económico.

- **Balanza comercial en dólares constantes**: Esta sería la diferencia entre las exportaciones e importaciones totales, ajustadas por inflación, para reflejar los valores en términos de poder adquisitivo de 2015. Al usar dólares constantes, elimina el efecto de la inflación y permite una comparación más precisa a lo largo del tiempo.

- **Merchandise exports by the reporting economy (current US$)** - Esta definición es más específica ya que menciona explícitamente que las cifras son "por la economía informante". Esto puede ser útil en análisis donde se necesite claridad sobre el origen de los datos y asegurar que las cifras son directamente reportadas y responsabilidad de la economía en cuestión.

- **Merchandise imports (current US$)** - Esta versión es más genérica y no especifica que los datos son reportados por la economía que los exporta o importa. Puede ser adecuada en contextos donde esta especificación no es necesaria o se entiende implícitamente.

# Balances Comerciales | Trade = X - M

```python
# Identificar columnas desde 1960 a 2020 en ambos dataframes
year_columns = [str(year) for year in range(1960, 2021)]

# Operación Aritmética
trade = exports[year_columns] - imports[year_columns]

# Incluir columnas 'Country Name' y 'Country Code' para identificarlos en el nuevo dataframe
trade['Country Name'] = exports['Country Name']
trade['Country Code'] = exports['Country Code']

# Reordenando columnas
trade = trade[['Country Name', 'Country Code'] + year_columns]

trade.head()
```

### Imputación con Mediana a valores NaN

```python
print(trade.isna().sum())
```

```python
import pandas as pd

# Crear una copia para trabajar con la imputación
trade_imputed = trade.copy()

# Identificar las columnas problemáticas
non_positive_columns = trade_imputed.loc[:, trade_imputed.columns[2:]].apply(lambda x: (x <= 0).any())
problematic_columns = non_positive_columns[non_positive_columns].index.tolist()

# Establecer un valor positivo mínimo para imputar
min_positive_value = 0.01

# Imputar valores no positivos
for column in problematic_columns:
    trade_imputed[column] = trade_imputed[column].apply(lambda x: min_positive_value if x <= 0 else x)

assert all(trade_imputed[problematic_columns].min() > 0)
```

```python
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.stattools import adfuller

# Imputar con Mediana
for column in trade_imputed.columns[2:]:
    median_value = trade_imputed[column].median()
    trade_imputed[column] = trade_imputed[column].fillna(median_value)

# Estandarizar los datos
scaler = StandardScaler()
trade_imputed.iloc[:, 2:] = scaler.fit_transform(trade_imputed.iloc[:, 2:])

# Estandarizar los datos
adf_result = adfuller(trade_imputed[column])
adf_result

trade_imputed.head(), adf_result
```

### World | Countries | Regions

```python
import pandas as pd

# Cargar el DataFrame principal
main_trade = pd.read_csv('main_trade.csv')

regions_list = [
    "Africa Eastern and Southern", "Africa Western and Central",
    "Central Europe and the Baltics", "East Asia & Pacific",
    "Europe & Central Asia", "European Union", "Latin America & Caribbean",
    "Middle East & North Africa", "North America", "OECD members",
    "Sub-Saharan Africa (excluding high income)", "South Asia (IDA & IBRD)"
]

# 1) DataFrame por países
trade_countries = main_trade[~main_trade['Country Name'].isin(regions_list + ['World'])]

# 2) DataFrame por regiones
trade_regions = main_trade[main_trade['Country Name'].isin(regions_list)]

# 3) DataFrame del mundo
trade_world = main_trade[main_trade['Country Name'] == 'World']

print(trade_countries.head())
print(trade_regions.head())
print(trade_world.head())
```

### Visualización de Tendencias Temporales | Countries

```python
import matplotlib.pyplot as plt
import seaborn as sns

countries_to_plot = ['United States', 'Germany', 'Brazil']
data_to_plot = trade_countries[trade_countries['Country Name'].isin(countries_to_plot)]

plt.figure(figsize=(20, 12))
for country in countries_to_plot:
    subset = data_to_plot[data_to_plot['Country Name'] == country]
    plt.plot(subset.columns[2:], subset.iloc[0, 2:], label=country)  

plt.title('Final Trade Balance in Constant Dollars Over Time')
plt.xlabel('Year')
plt.ylabel('Standardized Trade Balance')
plt.legend()
plt.grid(True)
plt.show()
```

### Comparación entre Regiones

```python
# Siempre elijo entre los años 2014 al 2020
regions_data_2020 = trade_regions[['Country Name', '2020']]  

plt.figure(figsize=(12, 8))
sns.barplot(x='2020', y='Country Name', data=regions_data_2020)
plt.title('Trade Balance in Constant Dollars by Region in 2020')
plt.xlabel('Standardized Trade Balance')
plt.ylabel('Region')
plt.show()
```

# Será lento

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

years = [str(year) for year in range(2014, 2021)]
trade_visual = trade_regions.set_index('Country Name')[years].transpose()

# Resumen estadístico
print(trade_visual.describe())

# Gráficos de series temporales para cada país
for column in trade_visual.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(trade_visual.index, trade_visual[column])
    plt.title(f'Trade Balance in Constant Dollars for {column}')
    plt.xlabel('Año')
    plt.ylabel('Final Trade Balance')
    plt.grid(True)
    plt.show()

# Descomposición de series temporales del total agregado
result = seasonal_decompose(trade_visual.sum(axis=1), model='additive', period=1)
result.plot()
plt.show()

# Correlación entre regiones
plt.figure(figsize=(10, 8))
sns.heatmap(trade_visual.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación entre Regiones')
plt.show()
```
